```python
## Exporting the Metadata on the Parquet file containing the Prompt Outputs

from cogsci.activations.importer import ExperimentDataImporter
import time
import os

dfs_metadata = []
for n_experiment in range(10):
    importer = ExperimentDataImporter(layer_name="output-block", n_experiment=n_experiment)
    csv_path = f"{importer.path_activations}/{importer.model_name}/gathered/metadata/metadata_experiment_{importer.n_experiment}.csv"

    if not os.path.exists(csv_path):
        start_time = time.time()  # Start timing before the process starts
        print(f"DOING EXPERIMENT {n_experiment}")
        df_metadata = importer.load_df_metadata()
        dfs_metadata.append(df_metadata)
        df_metadata.to_csv(csv_path)
        end_time = time.time()  # End timing after the process ends
        print(f"Time taken for experiment {n_experiment}: {end_time - start_time:.2f} seconds")  # Print the time taken
    else:
        print(f"CSV for experiment {n_experiment} already exists.")
```

```python
# Time taken: 1908.429224729538 seconds
# Exporting the Gathered Version of Activations

import os
from cogsci.activations.importer import ExperimentDataImporter

for n_experiment in range(10):

  print(n_experiment)

  # Mount the drive at the start of each iteration to handle potential disconnects
  drive.mount('/content/drive', force_remount=True)

  importer = ExperimentDataImporter(layer_name="output-block", n_experiment=n_experiment)
  path = f"{importer.path_activations}/{importer.model_name}/gathered/"
  filename = f"tmp={importer.n_experiment}_layer={importer.layer_name}_model={importer.model_name}_t=0.parquet"
  full_path = os.path.join(path, filename)

  if os.path.exists(full_path):
    print(f"file {full_path} already exists")
    continue

  try:
    df = importer.import_from_parquets()
    print(f"parquets imported into Memory for experiment {n_experiment}, now write them to single parquet to {full_path}")
    df.to_parquet(full_path)
  except OSError as e:
    if e.errno == 107:
      print(f"Transport endpoint is not connected for experiment {n_experiment}. Continuing...")
      drive.mount('/content/drive', force_remount=True)
      continue
    else:
      raise
```
